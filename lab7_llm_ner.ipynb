````python
import json
import re

def extract_json_from_response(text):
    match = re.search(r"```json(.*?)```", text, re.DOTALL)
    if not match:
        match = re.search(r"```(.*?)```", text, re.DOTALL)
    
    if match:
        json_text = match.group(1).strip()
        try:
            return json.loads(json_text)
        except json.JSONDecodeError:
            pass

    return {}
````

````python
base_command = """
{text}

Znajdź w powyższym tekście zwroty, które należą do każdej z kategorii:  
- data (np. 11 X 2012, 26 listopada 1914, maj 2010, 1410, 3 stycznia)
- godzina (np. 12:30, 8 rano, 4 po południu) 
- nazwa_geograficzna - nazwy kontynentów, regionów, pasm górskich, akwenów wodnych, wysp, półwyspów, zatok itp. (np. Mazury, Tatry, Półwysep Arabski, Morze Śródziemne, Zatoka Meksykańska) 
- nazwa_administracyjna - nazwy krajów, miast, wsi, ulic, dzielnic itp. (np. USA, Polska, Warszawa, Tarnów, Berlin, Skawina, Ulica Piotrkowska)
- osoba - imiona i nazwiska konkretnych osób, w tym zagraniczne (np. Maria Skłodowska-Curie, Akira Kurosawa, Donald Trump, Gerd Müller)
- organizacja - firmy, szkoły, uczelnie, instytucje państwowe, związki sportowe, organizacja pożytku publicznego itp. (np. Facebook, Inpost, AΙiExpress, Polski Związek Piłki Nożnej, Partia Pracy)
odpowiedź przedstaw w formacie JSON.
"""
````

````python
categories = ["data", "godzina", "nazwa_geograficzna", "nazwa_administracyjna", "osoba", "organizacja"]
````

````python
# as in previous lab

from datasets import load_dataset
import random

random.seed(42)

ds = load_dataset("clarin-knext/fiqa-pl", "corpus")
n = 1000
ds_sample = random.sample(ds['corpus']['text'], n)
````

````python
ds_sample[:3]
````

````python
import time
import json
from tqdm import tqdm

results = {category: [] for category in categories}

start_time = time.time()
i = 0

for sample in tqdm(ds_sample, desc="Processing prompts"):
    llm_result = run_with_bielik(base_command.format(text=sample))
    entity_dct = extract_json_from_response(llm_result)
    
    for category in categories:
        entities = entity_dct.get(category, [])
        for ent in entities:
            results[category].append(ent)

    i+=1
    if not i%50:
        with open(f'saved/lab7/llm_ner_single_shot_{i}.json', 'w') as file:
            json.dump(results, file, indent=4)

end_time = time.time()
runtime = round(end_time - start_time, 2)

with open('saved/lab7/llm_ner_single_shot.json', 'w') as file:
    json.dump(results, file, indent=4)
    
print(f"Time: {runtime} seconds")
````

````python
from collections import Counter

counts = {}

for entity_type in categories:
    found_of_type = results[entity_type]
    counts[entity_type] = len(found_of_type)
    freqs = Counter(found_of_type)
    print(f"\n_____ {entity_type} _____")
    print(f"      total: {counts[entity_type]}\n")
    for phrase, num in freqs.most_common(10):
        print(f"{phrase}: {num}")
````

````python
import spacy

nlp = spacy.load("pl_core_news_lg")

labels_transl = {
    'date': "data", 
    'geogName': "nazwa_geograficzna", 
    'orgName': "organizacja", 
    'persName': "osoba", 
    'placeName': "nazwa_administracyjna", 
    'time': "godzina",
}

def find_spacy_ners(nlp, text):
    text_nlp = nlp(text)
    lemmatized_text = ' '.join([token.lemma_ for token in text_nlp])
    lemma_text_nlp = nlp(lemmatized_text)
    ners = lemma_text_nlp.ents
    return [f"{ner.text} - {labels_transl[ner.label_]}" for ner in ners]
````

````python
def get_true_labels(labeled_text):
    result = []
    for label in labeled_text.keys():
        for item in labeled_text[label]:
            result.append(f"{item} - {label}")
    return result
````

````python
def find_bielik_zero_shot_ners(text):
    llm_result = run_with_bielik(base_command.format(text=text))
    entity_dct = extract_json_from_response(llm_result)
    result = []
    for label in entity_dct.keys():
        for item in entity_dct[label]:
            result.append(f"{item} - {label}")
    return result
````

````python
def find_bielik_few_shot_ners(text):
    llm_result = run_with_bielik(base_command_few_shot_v3 + text + "\n\nwynik: \n")
    entity_dct = extract_json_from_response(llm_result)
    result = []
    for label in entity_dct.keys():
        for item in entity_dct[label]:
            result.append(f"{item} - {label}")
    return result
````

````python
true_l = []
spacy_l = []
zero_shot_l = []
few_shot_l = []

for text in labeled_texts:
    true_labels = get_true_labels(text["result"])
    true_l.extend(true_labels)
    true_labels.sort()
    spacy_labels = find_spacy_ners(nlp, text["text"])
    spacy_l.extend(spacy_labels)
    spacy_labels.sort()
    zero_shot_labels = find_bielik_zero_shot_ners(text["text"])
    zero_shot_l.extend(zero_shot_labels)
    zero_shot_labels.sort()
    few_shot_labels = find_bielik_few_shot_ners(text["text"])
    few_shot_l.extend(few_shot_labels)
    few_shot_labels.sort()

    from itertools import zip_longest
    
    print(text["text"])
    print('\n')
    print("{:<30} | {:<30} | {:<30} | {:<30}".format("True", "SpaCy", "Bielik zero shot", "Bielik few shot"))
    print('-' * 129)
    for a, b, c, d in zip_longest(true_labels, spacy_labels, zero_shot_labels, few_shot_labels, fillvalue=""):
        print("{:<30} | {:<30} | {:<30} | {:<30}".format(a[:30], b[:30], c[:30], d[:30]))
        
    print('\n\n')
````

````python
def calc_metrics(result_set, model_set):
    true_positives = len(result_set & model_set)
    false_positives = len(model_set - result_set)
    false_negatives = len(result_set - model_set)
    
    precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return [precision, recall, f1]
````

````python
true_s = set(true_l)
spacy_s = set(spacy_l)
zero_shot_s = set(zero_shot_l)
few_shot_s = set(few_shot_l)

labels = ['precision', 'recall', 'F1']
spacy_m = calc_metrics(true_s, spacy_s)
zero_shot_m = calc_metrics(true_s, zero_shot_s)
few_shot_m = calc_metrics(true_s, few_shot_s)

print("{:<10} | {:<10} | {:<20} | {:<30}".format("", "SpaCy", "Bielik zero shot", "Bielik few shot"))
print('-' * 129)
for a, b, c, d in zip_longest(labels, spacy_m, zero_shot_m, few_shot_m):
    print("{:<10} | {:.3f}      | {:.3f}                | {:.3f}".format(a, b, c, d))
````

## Wnioski

Zastosowanie LLM do znajdywania NER wymaga znacznie większych zasobów, niż przetwarzanie NER za pomocą rozwiązania z biblioteki Spacy. Bez dostępu do nowoczesnej karty graficznej z dużą ilością pamięci wykonanie tego zadania nie jest możliwe. Zwiększa się przez to koszt każdego wyszukania encji, oraz jego czas. Na karcie użytej w powyższym badaniu - RTX 3060 12GB - na otrzymanie wyniku z modelu trzeba było każdorazowo czekać 3-4 sekundy (dla dłuższych komend z few shot learning nieco więcej). Wśród zalet podejście opertego o LLM należy jednak bez wątpienia wymienić wyższą skuteczność w rozpoznawaniu NER oraz większą elastyczność - chcąc zmienić/dodać kategorie nie trzeba trenować od nowa modelu - wystarczy zmienić wykorzystywany prompt.

W przeprowadzonym badaniu podejście oparte na LLM okazało się lepsze niż rozwiązanie ze SpaCy. Wydaje się jednak, że wyniki SpaCy są nieco zaniżone, ponieważ dla uproszczenia porównywane były tylko dokładne dopasowania. Przyglądając się wynikom Spacy można zauważyć wiele sytuacji, w których zakres wybrenego tekstu był nieco inny niż zakres wyznaczonej ręcznie labelki, ale również poprawny (np. "1957 - data" i "1957 rok - data", "16:00 - godzina" i "godzina 13:00 - godzina", "15 grudnia - data" i "15 grudzień - data"). Spacy ma jednak również większą tendencję do rozbijania całych encji na mniejsze części (np data "31.03.2025 " rozbita na "31. - godzina" (XD), "03. - godzina" i "2025 - data", czy "John J. Hopfield" podzielony na "John J." i "Hopfield").

Bardziej efektywną strategią promptowania okazało się zastosowanie few shot learningu. Próby rozwiązania tego zagadnienia za pomocą zero shot learning pododowały, że niekiedy model wśród wyznaczonych encji zwracał elementy, które należały do przykładów kategorii podanych w prompcie. Warto dodać, że Bielik w obydwu sytuacjach nie miał problemów ze zwracaniem rezultatów w pożądanym formacie JSON, co ułatwiało ich późniejsze parsowanie. Skuteczność modelu językowego w takim zastosowaniu w dużej mierze opiera się na tym, jaki prompt zostanie użyty - źle dobrane polecenie może wprowadzić dodatkowe problemy, na przykład sprawiając, że model będzie znajdywał encje, których nie ma w tekście, a które zostały mu przekazane jako przykłady. Warto również doprecyzować kategorie, na których model ma działać (problemy sprawiało np. oddzielenie nazw geograficznych od nazw administracyjnych).

Zastosowanie tradycyjnego NER jest zdecydowanie rekomendowane wszędzie tam, gdzie konieczne jest radzenie sobie z bardziej ograniczonymi zasobami (np. na urządzeniach mobilnych), lub konieczna jest szubka odpowiedź (3-4 sekundy oczekiwania na odpowiedź to jednak dość dużo).
